```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head(3)
```

```python
df.info()
```

```python
del df["street"]
del df["date"]
del df['country']
```

```python
print(len(df['statezip'].unique()))
print(len(df['city'].unique()))
```

```python
from sklearn.preprocessing import OneHotEncoder
transform = OneHotEncoder(drop="first", sparse_output=False)
df['statezip'] = transform.fit_transform(df[['statezip']])
df[ 'city'] = transform.fit_transform(df[['city']])
```

```python
from sklearn.preprocessing import LabelEncoder
lable_encode = LabelEncoder()
df["price"] =  lable_encode.fit_transform(df["price"])
y = df["price"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['scaled_price'] = scaler.fit_transform(df[['price']])
y = df['scaled_price']
```

```python
x = df.iloc[:, 1:15]
x.head()
```

```python
print(x.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# from sklearn.ensemble import RandomForestRegressor
# model = RandomForestRegressor(n_estimators=255, random_state=42)
# model.fit(x_train, y_train)

from sklearn.linear_model import Ridge
model = Ridge(alpha=10)
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rmse = np.sqrt(mean_squared_error(y_test, pred))
print(f"Random Forest RMSE: {rmse:.2f}")
print(f"Random Forest R2_score: {r2_score(y_test, pred):.4f}")
```
